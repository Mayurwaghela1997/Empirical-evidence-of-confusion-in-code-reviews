```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import squarify 
import seaborn as sns
from scipy import stats

plt.rcParams["figure.figsize"] = [7.50, 7]
plt.rcParams["figure.autolayout"] = True
```

```python
#What are the coding conventions if not followed by the author of change will confuse you as a reviewer 
Q9_df = pd.read_excel('Survey_Q9_answer.xls').iloc[:,2:]
Q9_df = Q9_df.dropna(how='all')
Q9_data = Q9_df.iloc[:,0].append(Q9_df.iloc[:,1]).append(Q9_df.iloc[:,2]).append(Q9_df.iloc[:,3]).append(Q9_df.iloc[:,4]).dropna()
```

```python
fig, ax = plt.subplots()
Q9_data.value_counts().sort_values().plot(ax=ax, kind='barh', xlabel='Coding conventions', ylabel='frequency from 21 responses')
# plt.xticks(np.arange(0, 10, 1.0))
plt.rcParams["figure.figsize"] = [9, 7]
plt.title('Coding conventions that if not followed by author of change will confuse the reviewers \n')
plt.show()
```

```python
#if not code smell then what was the reason for your confusion
Q8_data = pd.read_excel('Survey_Q8_answer.xls').iloc[:,2].dropna()
sizes=[25, 35, 52, 35,35,25]
label= Q8_data
color=['red','blue','green','grey', 'yellow', 'purple']
squarify.plot(sizes=sizes, label=label, color=color, alpha=0.6 )
plt.axis('off')
plt.title('Reasons of confusion other than code smells')
plt.show()
```

```python
confusion_smell = data[data['Num_label']==1]
sample_smell = data[data['Num_label']==0]
data1 = confusion_smell['Smell_added']
data2 = sample_smell['Smell_added']


data1DF = pd.DataFrame(columns=['type', 'added_smell_count'])
data1DF['added_smell_count'] = data1

data2DF = pd.DataFrame(columns=['type', 'added_smell_count'])
data2DF['added_smell_count'] = data2
df = data1DF.fillna('Confusion').append(data2DF.fillna('No Confusion'))


ax = df.boxplot(column='added_smell_count', by='type', showfliers=True, 
                positions=range(df.type.unique().shape[0]), figsize =(10, 7))

sns.pointplot(x='type', y='added_smell_count', data=df.groupby('type', as_index=False).median(), ax=ax)
plt.ylabel('No. of code smells added')
plt.xlabel('')
ax.yaxis.grid(True) # Hide the horizontal gridlines
ax.xaxis.grid(True) # Show the vertical gridlines
plt.title('Correlation between confusion & No confusion with No. of code smells added')
plt.show()
```
